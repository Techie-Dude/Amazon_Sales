# **<p style="font-family:lora;font-size:100%;text-align:center;color:#000000;">Changing Data Types of Columns from object to float</p>**

```python
df[['discounted_price','actual_price','discount_percentage','rating','rating_count']]
```

```python
df['discounted_price'] = df['discounted_price'].str.replace("â‚¹",'')
df['discounted_price'] = df['discounted_price'].str.replace(",",'')
df['discounted_price'] = df['discounted_price'].astype('float64')

df['actual_price'] = df['actual_price'].str.replace("â‚¹",'')
df['actual_price'] = df['actual_price'].str.replace(",",'')
df['actual_price'] = df['actual_price'].astype('float64')

df['discount_percentage'] = df['discount_percentage'].str.replace('%','').astype('float64')
df['discount_percentage'] = df['discount_percentage'] / 100

df['rating'] = df['rating'].astype('float64')

df['rating_count'] = df['rating_count'].str.replace(",",'',regex=True)
df['rating_count'] = df['rating_count'].astype('float64')
```

## ***Q1:Which products have the highest average ratings***

```python
top_rated_products = df.groupby('product_name')['rating'].mean().sort_values(ascending=False).head(10)
top_rated_products = pd.Series(top_rated_products)
def truncate_product_name(name):
    if '|' in name:
        return name.split('|')[0].strip()
    elif ',' in name:
        return name.split(',')[0].strip()
    elif '5/' in name:
        return name.split('/')[0].strip()
    else:
        return name

top_rated_products.index = top_rated_products.index.map(truncate_product_name)
top_rated_products
```

```python
plt.figure(figsize=(12, 6))
plt.bar(top_rated_products.index, top_rated_products, color='darkorange')
plt.title('Best Product Performance')
plt.xlabel('Product Name')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.show()
```

## ***Interference:Based on the given data, we can observe the following:***

## ***The top three products, which have the highest average ratings of 5.0, are the Amazon Basics Wireless Mouse, REDTECH USB-C to Lightning Cable, and Syncwire LTG to USB Cable.***
## ***Products related to electronics and household utilities such as the Oraletch Coffee Frother electric and the Instant Pot Air Fryer also have high ratings, close to 4.8.***
## ***The lowest average rating among the top ten products is 4.7, which still indicates a high level of customer satisfaction across these top-rated products.***
